```python
# Importing all necessary libraries
import cv2
import os
  
# Read the video from specified path
cam = cv2.VideoCapture("/Users/djhamb/Desktop/Tennis Videos/2tennis#2.mp4")
  
if not cam.isOpened():
    print("Error: Unable to open video file")
    exit()
    
currentframe=0

try:
      
    # creating a folder named data
    if not os.path.exists('2tennis#2'):
        os.makedirs('2tennis#2')
  
except OSError:
    print ('Error: Creating directory of TennisExtraction')
  
  
while(True):
    # reading from frame
    ret,frame = cam.read()
  
    if ret:
        # if video is still left continue creating images
        name = './2tennis#2/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name)
  
        # writing the extracted images
        cv2.imwrite(name, frame)
  
        # increasing counter so that it will
        # show how many frames are created
        currentframe += 1
    else:
        break
  
cam.release()
cv2.destroyAllWindows()
```
